## Add 3 New Columns to the below df
See [This Google Sheet](https://docs.google.com/spreadsheets/d/1f71aepuU2NvJI1EoFWpl3eCxypLFokJGNOdwIz8Xj1o/edit#gid=1038614599) for an example of how I would like these calculated
- same_gaps_last45
- same_gaps_covered60_last45
- same_gaps_covered60_last45_pct



```python
from datetime import datetime
import pandas as pd
import numpy as np

df = pd.read_csv ('Question 1 - Table.csv')

df = df[[
    'date',
    'symbol',
    'pos_type',
    'fut_gap_filled',
    ]]

df['date'] = pd.to_datetime(df['date'])

df
```

```python
#same_gaps_last45

timestamp = pd.Series([i.timestamp() for i in df['date']])
        
def same_gaps_last45(row): 
        
    if row['date'] != "":
        
        _ = df[
            (df['symbol'] == row['symbol']) & 
            (df['pos_type'] == row['pos_type']) & 
            (row['date'].timestamp() > timestamp) &             
            (timestamp > row['date'].timestamp()-3888000.0)]
            
        return len(_)

df["same_gaps_last45"] = df.apply(lambda row: same_gaps_last45(row), axis=1)

df
```

```python
#same_gaps_covered60_last45

timestamp = pd.Series([i.timestamp() for i in df['date']])

def same_gaps_covered60_last45(row):
    
    if row['date'] != "":
        
        _ = df[
            (df['symbol'] == row['symbol']) & 
            (df['pos_type'] == row['pos_type']) & 
            (df['fut_gap_filled'] >= 0.6) & 
            (row['date'].timestamp() > timestamp) & 
            (timestamp > row['date'].timestamp()-3888000.0)]
        
        return len(_)

df["same_gaps_covered60_last45"] = df.apply(lambda row: same_gaps_covered60_last45(row), axis=1)

df
```

```python
#same_gaps_covered60_last45_pct

def same_gaps_covered60_last45_pct(row):
    
     if row['date'] != "":
            try:
                return int(row['same_gaps_covered60_last45'] / row['same_gaps_last45'])
            except ZeroDivisionError:
                return np.nan
            
df["same_gaps_covered60_last45_pct"] = df.apply(lambda row: same_gaps_covered60_last45_pct(row), axis=1)

df
```
