## Reduce X amount of Duplicates based on Multiple Criteria
See [This Google Sheet](https://docs.google.com/spreadsheets/d/1f71aepuU2NvJI1EoFWpl3eCxypLFokJGNOdwIz8Xj1o/edit#gid=1385175813) for an example of what I would like
1. Sort df by {date} ASC then by {gap} DESC then by {now_open} ASC
2. Count the Unique Occurences of the {date} values ({unique_count})
3. Create a new df (df2) that excludes all {unique_count} > 10


```python
import pandas as pd

df = pd.read_csv ('Question 2 - Table.csv')

df = df[[
    'date',
    'symbol',
    'gap',
    'pos_type',
    'now_open',
    ]]

df
```

```python
#Sort df by {date} ASC then by {gap} DESC then by {now_open} ASC

df = df.sort_values(by=['date', 'gap', 'now_open'], ascending=[True, False, True])

df
```

```python
#Count the Unique Occurences of the {date} values ({unique_count})

lst = []
uniques = df.value_counts('date')

for unique in uniques: 
    lst += list(range(1, unique+1))

df["unique_count"] = pd.Series(lst)

df
```

```python
#Create a new df (df2) that excludes all {unique_count} > 10

dates = df.value_counts('date').keys().tolist()

for date in dates:
    temp = df.loc[df['date'] == date][10:]
    trash = temp.index.tolist()
    df2 = df.drop(trash)

df2 = df2.reset_index(drop = True)
    
df2
```
